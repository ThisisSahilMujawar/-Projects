<h3>Task 2: Created a function which can extract city name along with the country state</h3>

```python
def City_US(address):
    return address.split(',')[1].split(' ',maxsplit = 1)[1]
def State_US(address):
    return address.split(',')[2].split(' ')[1]

all_df['City'] = all_df['Purchase Address'].apply(lambda x: f"{City_US(x)} ({State_US(x)})")
```

```python
all_df['Time'] = all_df['Time'].astype(str)
all_df['Order Date'] = all_df['Order Date'].astype(str)
def get_time(date):
    return date.split(' ',maxsplit = 1)[1].rsplit(':',maxsplit = 1)[0]
all_df['Time'] = all_df['Order Date'].apply(lambda x: get_time(x))
all_df['Time']
```

```python
all_df['Time'][0]
```

```python
# all_df.drop(columns = ['Exact Time'])
```

```python
def get_min(time):
    s = int(time.split(':')[0])
    r = int(time.split(':')[1])
    if r> 30:
        return s+1
    else:
        return s

all_df['Exact Time'] = all_df['Time'].apply(lambda x: get_min(x))
```

```python
all_df.head()
```

```python
all_df['Exact Time'].dtype
```

```python
all_df.loc[all_df['Exact Time'] == 24,'Exact Time'] = 0
all_df.head()
all_df['Count'] = 1
all_df.head()
```

```python
all_df.groupby('Exact Time').sum()['Count']
```

```python
results3 = all_df.groupby('Exact Time').sum()['Count']
Hour = list(results3.index)
plt.bar(Hour, results3 )
plt.xticks(Hour, rotation = 'vertical', size = 8)
plt.ylabel('Best Time')
plt.xlabel('Hour')
plt.show()
```

<h3>Question 4 What products are most often sold together ?</h3>

```python
all_df.head()
```

```python
all_df['Order ID'].duplicated(keep = False)
```

```python
df = all_df[all_df['Order ID'].duplicated(keep = False)]
# df = df.drop(columns = df.columns.values[9:])
```

```python
df.head(10)
```

```python
# df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))    same step as below
# avoided warning signs using .loc:
df.loc[:, 'Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df
```

```python
df = df[['Order ID','Grouped']].drop_duplicates()
```

```python
df.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()
for row in df['Grouped']:    
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```
